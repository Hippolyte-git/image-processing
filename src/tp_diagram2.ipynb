```python
def crofton_perimeter(I):
    """ Computation of crofton perimeter """
    inter = []
    h = np.array([[1, -1]])
    for i in range(4):
        II = np.copy(I)
        # Rotate the image at 0, 45, 90, and 135 degrees
        I2 = rotate(II, 45 * i, reshape=True, order=0) > 0.5
        I3 = scipy.ndimage.convolve(I2, h)
        inter.append(np.sum(I3))
    
    # Crofton formula approximation with 4 orientations
    crofton = np.pi / 4. * (inter[0] + inter[2] + (inter[1] + inter[3]) / np.sqrt(2))
    return crofton

def feret_diameter(I):
    """
    Computation of the Feret diameter
    d: minimum Feret diameter (meso-diameter)
    D: maximum Feret diameter (exo-diameter)
    """
    d = np.max(I.shape)
    D = 0
    I_float = I.astype(float)
    
    # Evaluate projections every 30 degrees
    for a in np.arange(0, 180, 30):
        I2 = rotate(I_float, a, reshape=True, order=0) > 0.5
        F = np.max(I2, axis=0)
        measure = np.sum(F)
        
        if measure < d:
            d = measure
        if measure > D:
            D = measure
            
    return d, D

def inscribedRadius(I):
    """
    computes the radius of the inscribed circle
    """
    binary = I > 0.5
    dm = scipy.ndimage.distance_transform_cdt(binary)
    radius = np.max(dm)
    return radius

def circumCircle(I):
    """
    computes the center and radius of the circumscribed circle
    """
    points = np.argwhere(I > 0.5)
    c = smallestenclosingcircle.make_circle(points)
    return c
```
